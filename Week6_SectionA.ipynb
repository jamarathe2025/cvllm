# Section A: LLM Foundations & Hugging Face
**Hugging Face Setup & Text Generation**
1. Install Hugging Face transformers.
2. Load a small LLM (e.g., distilgpt2).
3. Generate three different continuations for the prompt: “AI is transforming industries by …”.

Deliverable: Notebook cell with outputs.

```python
pip install transformers torch
```

Setup 1

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="distilbert/distilgpt2")
```

```python
prompt = "AI is transforming industries by"
```

```python
results = generator(
    prompt,
    max_new_tokens=30,
    num_return_sequences=3,
    do_sample=True,
    temperature=0.5
)
```

```python
for i, res in enumerate(results):
    print(f"Response Continuation {i+1}: {res['generated_text']}\n")
```

Displaying output HTML table

```python
gen_rows = ""
for i, res in enumerate(results):
    text = res['generated_text']
    gen_rows += f"<tr><td style='padding:8px;'><b>Continuation {i+1}</b></td><td style='padding:8px;'>{text}</td></tr>"

gen_html = f"""
<h3>1. LLM Generation Results (DistilGPT2)</h3>
<p><b>Prompt:</b> {prompt}</p>
<table border="1" style="width:100%; border-collapse:collapse; font-family:sans-serif;">
    <tr style="background-color: #e6f3ff;">
        <th style="padding:10px; width:20%;">ID</th>
        <th style="padding:10px;">Generated Text</th>
    </tr>
    {gen_rows}
</table>
<br><hr>
"""
display(HTML(gen_html))
```

# **Section A: LLM Foundations & Hugging Face**
**Tokenization Demo**
1. Take the sentence: “LLMs are powerful tools for natural language understanding.”
2. Tokenise it using a Hugging Face tokeniser.
3. Display the resulting tokens, token IDs, and sequence length.

```python
from transformers import AutoTokenizer
```

Load the tokenizer for distilgpt2

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilgpt2")
```

Define the sentence

```python
sentence = "LLMs are powerful tools for natural language understanding."
```

Tokenize the sentence and Get individual tokens

```python
inputs = tokenizer(sentence)
tokens = tokenizer.tokenize(sentence)
```

Extract IDs and Sequence Length

```python
token_ids = inputs['input_ids']
sequence_length = len(token_ids)
```

Final result

```python
print(f"Sentence: {sentence}")
print(f"-"*30)
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")
print(f"Sequence Length: {sequence_length} tokens")
```

Making result more presentable using HTML table

```python
from IPython.display import HTML, display

# --- Replace your '# Display Results' section with this ---

# 1. Prepare table rows
table_rows = ""
for i, (token, tid) in enumerate(zip(tokens, token_ids)):
    # GPT tokenizers use 'Ġ' to represent a space; we replace it for readability
    readable_token = token.replace('Ġ', ' ')
    table_rows += f"<tr><td>{i}</td><td>{readable_token}</td><td>{tid}</td></tr>"

# 2. Define the HTML structure
html_table = f"""
<h3>Tokenization Analysis</h3>
<p><b>Sentence:</b> {sentence}</p>
<p><b>Total Sequence Length:</b> {len(token_ids)}</p>
<table border="1" style="width:50%; border-collapse:collapse; text-align:left;">
    <tr style="background-color: #f2f2f2;">
        <th>Index</th>
        <th>Token</th>
        <th>Token ID</th>
    </tr>
    {table_rows}
</table>
"""

# 3. Render the table in Colab
display(HTML(html_table))
```
