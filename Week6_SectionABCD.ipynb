# **Section A: LLM Foundations & Hugging Face**
**Tokenization Demo**
1. Take the sentence: “LLMs are powerful tools for natural language understanding.”
2. Tokenise it using a Hugging Face tokeniser.
3. Display the resulting tokens, token IDs, and sequence length.

```python
from transformers import AutoTokenizer
```

Load the tokenizer for distilgpt2

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilgpt2")
```

Define the sentence

```python
sentence = "LLMs are powerful tools for natural language understanding."
```

Tokenize the sentence and Get individual tokens

```python
inputs = tokenizer(sentence)
tokens = tokenizer.tokenize(sentence)
```

Extract IDs and Sequence Length

```python
token_ids = inputs['input_ids']
sequence_length = len(token_ids)
```

Final result

```python
print(f"Sentence: {sentence}")
print(f"-"*30)
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")
print(f"Sequence Length: {sequence_length} tokens")
```

Making result more presentable using HTML table

```python

# 1. Prepare table rows
table_rows = ""
for i, (token, tid) in enumerate(zip(tokens, token_ids)):
    # GPT tokenizers use 'Ġ' to represent a space; we replace it for readability
    readable_token = token.replace('Ġ', ' ')
    table_rows += f"<tr><td>{i}</td><td>{readable_token}</td><td>{tid}</td></tr>"

# 2. Define the HTML structure
html_table = f"""
<h3>Tokenization Analysis</h3>
<p><b>Sentence:</b> {sentence}</p>
<p><b>Total Sequence Length:</b> {len(token_ids)}</p>
<table border="1" style="width:50%; border-collapse:collapse; text-align:left;">
    <tr style="background-color: #f2f2f2;">
        <th>Index</th>
        <th>Token</th>
        <th>Token ID</th>
    </tr>
    {table_rows}
</table>
"""

# 3. Render the table in Colab
display(HTML(html_table))
```

Loading pipeline and HTML modules for better display

```python
from transformers import pipeline
from IPython.display import HTML, display
```

Initialize the generator

```python
generator = pipeline("text-generation", model="distilbert/distilgpt2")
```

Loading the Paragraph for summary

Build Task dictionay for 3 PROMPT tasks as per assignment

```python
tasks = [
    {
        "name": "Summarization",
        "prompt": "Summarize this in one sentence: Artificial intelligence is intelligence demonstrated by machines, as opposed to natural intelligence displayed by animals. AI research defines intelligent agents as systems that perceive their environment.",
        "tokens": 20
    },
    {
        "name": "Q&A",
        "prompt": "Question: What is the capital of India? Answer:",
        "tokens": 20
    },
    {
        "name": "Poem",
        "prompt": "Write a short 4-line poem about AI:",
        "tokens": 60
    }
]
```

```python
html_output = """
<style>
    .result-table { width: 100%; border-collapse: collapse; font-family: Arial; }
    .result-table th { background-color: #2c3e50; color: white; padding: 12px; }
    .result-table td { border: 1px solid #ddd; padding: 10px; vertical-align: top; width: 30%; }
    .task-header { background-color: #ecf0f1; font-weight: bold; }
</style>
<table class="result-table">
    <tr>
        <th>Task Description</th>
        <th>Variation 1</th>
        <th>Variation 2</th>
        <th>Variation 3</th>
    </tr>
"""
```

Generating results

```python
for t in tasks:
    print(f"Generating 3 variations for: {t['name']}...")
    results = generator(
        t['prompt'],
        max_new_tokens=t['tokens'],
        num_return_sequences=3,
        do_sample=True,
        temperature=0.7,
        truncation=True
    )

    # --- Debugging print for summarization ---
    if t['name'] == 'Summarization':
        print("--- Raw Generated Text for Summarization (before stripping prompt) ---")
        for i, res in enumerate(results):
            print(f"Variation {i+1}: {res['generated_text']}")
        print("--------------------------------------------------------------------")
    # --- Debugging print for Q&A ---
    if t['name'] == 'Q&A':
        print("--- Raw Generated Text for Q&A (before stripping prompt) ---")
        for i, res in enumerate(results):
            print(f"Variation {i+1}: {res['generated_text']}")
        print("--------------------------------------------------------------------")
    # --- End Debugging print ---

    # Extract only the NEW text (stripping the prompt)
    v1 = results[0]['generated_text'].replace(t['prompt'], "").strip()
    v2 = results[1]['generated_text'].replace(t['prompt'], "").strip()
    v3 = results[2]['generated_text'].replace(t['prompt'], "").strip()

    # Add a row to our table string
    html_output += f"""
    <tr>
        <td class="task-header">{t['name']}</td>
        <td>{v1 if v1 else "<i>(No output)</i>"}</td>
        <td>{v2 if v2 else "<i>(No output)</i>"}</td>
        <td>{v3 if v3 else "<i>(No output)</i>"}</td>
    </tr>
    """
```

Show HTML formatted result

```python
html_output += "</table>"
display(HTML(html_output))
```

# Section C: Embeddings with Gensim
## **Word Embeddings**
1. Use GloVe embeddings (glove-wiki-gigaword-50) via Gensim.
2. Select any three words (e.g., king, queen, diamond).
3. For each word, display:
First 10 values of its vector.
Top 5 most similar words and their similarity scores

```python
pip install -q gensim
```

```python
import gensim.downloader as api
from IPython.display import HTML, display
```

```python
# 2. Download the glove-wiki-gigaword-50 model (approx. 66MB)
model = api.load("glove-wiki-gigaword-50")
```

```python
# 3. Select words and prepare data
words = ["king", "queen", "diamond"]
html_output = """
<style>
    .emb-table { width: 100%; border-collapse: collapse; font-family: sans-serif; }
    .emb-table th { background-color: #27ae60; color: white; padding: 10px; }
    .emb-table td { border: 1px solid #ddd; padding: 10px; vertical-align: top; }
    .vector-box { font-family: monospace; font-size: 12px; background: #f4f4f4; padding: 5px; border-radius: 4px; }
</style>
<table class="emb-table">
    <tr>
        <th>Word</th>
        <th>First 10 Vector Values</th>
        <th>Top 5 Similar Words (Score)</th>
    </tr>
"""
```

```python
for word in words:
    # Get the word vector and first 10 values
    vector = model[word][:10]
    vector_str = ", ".join([f"{val:.4f}" for val in vector])

    # Get top 5 most similar words
    similar = model.most_similar(word, topn=5)
    similar_str = "<br>".join([f"{w} ({score:.4f})" for w, score in similar])

    html_output += f"""
    <tr>
        <td><b>{word}</b></td>
        <td><div class="vector-box">{vector_str}...</div></td>
        <td>{similar_str}</td>
    </tr>
    """
```

```python
html_output += "</table>"
display(HTML(html_output))
```

# Section C: Embeddings with Gensim
## **Sentence-Level Embeddings**
1. Create five short sentences (e.g., on AI or jewellery).
2. Compute the average of word vectors for each sentence (a simple baseline for sentence representation).
3. Show a similarity matrix using cosine similarity to compare the sentences.

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import HTML, display
```

```python
# 1. Define 5 sentences (Mixed AI and Jewellery themes)
sentences = [
    "AI is transforming the diamond industry.",
    "Machine learning helps design gold rings.",
    "Artificial intelligence creates smart software.",
    "A diamond necklace is a luxury gift.",
    "Generative models can write poetry."
]
```

```python
def get_sentence_vector(sentence, model):
    # Split into lowercase words and filter those not in GloVe vocabulary
    words = [w.lower() for w in sentence.replace(".", "").split() if w.lower() in model]
    if not words:
        return np.zeros(50)
    # Calculate the average (mean) of all word vectors in the sentence
    word_vectors = [model[w] for w in words]
    return np.mean(word_vectors, axis=0)
```

```python
# 2. Compute vectors for each sentence
sentence_vectors = np.array([get_sentence_vector(s, model) for s in sentences])
```

```python
# 3. Calculate Cosine Similarity Matrix
# Result is a 5x5 matrix comparing every sentence to every other sentence
sim_matrix = cosine_similarity(sentence_vectors)
```

```python
# 4. Display as a Stylized HTML Table
df = pd.DataFrame(sim_matrix, columns=[f"S{i+1}" for i in range(5)], index=[f"S{i+1}" for i in range(5)])

html_table = "<h3>Sentence Similarity Matrix (Cosine Similarity)</h3>"
html_table += "<p><b>S1:</b> " + sentences[0] + " | <b>S2:</b> " + sentences[1] + " | <b>S3:</b> " + sentences[2] + "<br>"
html_table += "<b>S4:</b> " + sentences[3] + " | <b>S5:</b> " + sentences[4] + "</p>"
```

```python
# Use Pandas styling to add a heatmap effect
styled_df = df.style.background_gradient(cmap='YlGnBu', axis=None).format("{:.2f}")
display(HTML(html_table))
display(styled_df)
```

# Sentiment Analysis Demo

```python
from transformers import pipeline
from IPython.display import HTML, display
```

```python
# 1. Initialize the Sentiment Analysis pipeline
# This model classifies text as POSITIVE or NEGATIVE
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# 2. Define custom inputs (Mix of positive, negative, and neutral-ish sentences)
custom_inputs = [
    "The new AI-powered diamond sorting tool is incredibly efficient and fast!",
    "I'm quite disappointed with the lag in the latest software update.",
    "The necklace arrived, but the packaging was damaged and looked cheap.",
    "The fusion of technology and traditional craftsmanship is the future of luxury.",
    "It's an okay product, but definitely not worth the premium price tag."
]
```

```python
# 3. Run Inference
results = classifier(custom_inputs)
```

```python
# 4. Build a Stylized HTML Table to display results
html_table = """
<style>
    .sent-table { width: 100%; border-collapse: collapse; font-family: sans-serif; margin-top: 20px; }
    .sent-table th { background-color: #34495e; color: white; padding: 12px; text-align: left; }
    .sent-table td { border: 1px solid #ddd; padding: 12px; }
    .pos { color: #27ae60; font-weight: bold; } /* Green for Positive */
    .neg { color: #e74c3c; font-weight: bold; } /* Red for Negative */
    .score-bar { background: #eee; border-radius: 4px; height: 10px; width: 100px; display: inline-block; margin-left: 10px; }
    .score-fill { background: #3498db; height: 10px; border-radius: 4px; }
</style>
<table class="sent-table">
    <tr>
        <th>Input Text</th>
        <th>Sentiment Label</th>
        <th>Confidence Score</th>
    </tr>
"""
```

```python
for text, res in zip(custom_inputs, results):
    label_class = "pos" if res['label'] == "POSITIVE" else "neg"
    confidence_pct = res['score'] * 100

    html_table += f"""
    <tr>
        <td>{text}</td>
        <td class="{label_class}">{res['label']}</td>
        <td>
            {res['score']:.4f}
            <div class="score-bar"><div class="score-fill" style="width: {confidence_pct}%;"></div></div>
        </td>
    </tr>
    """
```

```python
html_table += "</table>"
display(HTML(html_table))
```

# Translation Demo English to Hindi

```python
# 1. Install necessary dependencies for Hindi translation
!pip install -q transformers[sentencepiece] torch
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# 1. Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
```

```python
# 2. Define custom inputs for translation
custom_inputs_translation = [
    "Hello, how are you?",
    "Artificial intelligence is changing the world.",
    "The diamond industry is embracing new technologies.",
    "Colab is a great tool for machine learning."
]
```

```python
# 3. Perform translation for each input
translated_results = []
for text in custom_inputs_translation:
    # Prepare the input for the model
    input_ids = tokenizer(text, return_tensors="pt").input_ids

    # Generate the translation
    # Using num_beams for better quality translation
    translated_tokens = model.generate(input_ids, num_beams=5, max_length=128)

    # Decode the tokens back to text
    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    translated_results.append({"original": text, "translated": translated_text})
```

```python
# 4. Display results in a stylized HTML table
html_translation_output = """
<style>
    .trans-table { width: 100%; border-collapse: collapse; font-family: sans-serif; margin-top: 20px; }
    .trans-table th { background-color: #f39c12; color: white; padding: 12px; text-align: left; }
    .trans-table td { border: 1px solid #ddd; padding: 12px; }
</style>
<table class="trans-table">
    <tr>
        <th>Original English Text</th>
        <th>Translated Hindi Text</th>
    </tr>
"""

for res in translated_results:
    html_translation_output += f"""
    <tr>
        <td>{res['original']}</td>
        <td>{res['translated']}</td>
    </tr>
    """

html_translation_output += "</table>"
display(HTML(html_translation_output))
```
